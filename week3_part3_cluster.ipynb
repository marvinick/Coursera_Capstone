```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df_content = requests.get(page).text
df_parse = BeautifulSoup(df_content, 'lxml')
df_postcodes = df_parse.find('table', class_='wikitable')

```

```python
first_iteration = True 
for tablerow in df_postcodes.find_all('tr'):
    if first_iteration:
        #initialize dataframe 
        #column name as the first entry for the table 
        col_names = tablerow.text.split('\n')[1:-1]
        postal_codes = pd.DataFrame(columns=col_names)
        first_iteration = False 
    elif tablerow.text.split('\n')[2] == 'Not assigned':
        #ignore if it's not assigned 
        continue
    elif tablerow.text.split('\n')[3] == 'Not assigned':
        #name of the burrough if neighborhood is NA 
        row = tablerow.text.split('\n')[1:-1]
        row[-1] = row[1]
        postal_codes = postal_codes.append({'Postcode':row[0],'Borough':row[1],'Neighbourhood':row[2]},ignore_index=True)
    else:
        postal_row = tablerow.text.split('\n')[1:-1] 
        postal_codes = postal_codes.append({'Postcode':postal_row[0],'Borough':postal_row[1],'Neighbourhood':postal_row[2]},ignore_index=True)
        
```

```python
postal_codes = postal_codes.groupby(['Postcode','Borough'],as_index=False, sort=False).Neighbourhood.agg(', '.join)
postal_codes
```

```python
postal_codes.shape[0]
```

```python
geo_csv = pd.read_csv('http://cocl.us/Geospatial_data')
geo_csv
```

```python
geo_csv.columns = ['Postcode', 'Latitude', 'Longitude']
geo_csv.columns 
postal_codes = postal_codes.merge(geo_csv, on='Postcode')
postal_codes
```

```python
import folium
# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
# import k-means from clustering stage
from sklearn.cluster import KMeans
from geopy.geocoders import Nominatim
```

```python
address = 'Toronto, Canada'
geolocator = Nominatim()
location = geolocator.geocode(address)
lat = location.latitude 
long = location.longitude 
print('Toronto coordinates are {}, {}.'.format(lat, long))
```

```python
toronto_map = folium.Map(location=[lat, long], zoom_start=10)

#markers 
for lat, long, borough, neighbourhood in zip(postal_codes['Latitude'], postal_codes['Longitude'], postal_codes['Borough'], postal_codes['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood,borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker([lat, long], radius=5, popup=label, 
                        color='blue',fill=True, fill_color='#3186cc',
                        fill_opacity=0.7).add_to(toronto_map)

toronto_map
    
```
